```python
def getBreweries_byState(State='None'):
        if State=='None':
            #json_string=Breweries_df.to_json()
            json_string=Breweries_df.sort_values(by=['state']).to_json() 
        else: 
            State=State.upper()
            json_string =Breweries_df.loc[Breweries_df['state']==State].to_json()
        return(json_string)  
 
```

```python
def getBreweries_byCity(City='None'):
        if City=='None':
            #json_string=Breweries_df.to_json()
            json_string=Breweries_df.sort_values(by=['city']).to_json()    
        else: 
            City=City.title()
            json_string =Breweries_df.loc[Breweries_df['city']==City].to_json()
        return(json_string)  
```

```python
def getBreweries_byType(Type='None'):
        if Type=='None':
            #json_string=Breweries_df.to_json()
            json_string='Please supply a beer type: Ex.. Lager, Stout, hefeweizen  '   
        else: 
            Type=Type.lower()
            json_string =Merged_df.sort_values(by=['state']).loc[Merged_df['beer_style']==Type].to_json()
        return(json_string)  
```

### FLASK

```python
app = Flask(__name__)

@app.route("/")
def index():
      routes='''
      <p>
      API end points for search breweries and beers.  
      </p>
      Routes:
            <ul>
            <li><a href="/api/v1.0/Breweries">/api/v1.0/Breweries</a>
            <li>/api/v1.0/Cities/&ltCityName&gt
            <li><a href="/api/v1.0/State">/api/v1.0/State</a>
            <li>/api/v1.0/State/&ltstateCode&gt    
            <li>/api/v1.0/Beers/&ltbyType&gt
            </ul>'''   
      return routes

@app.route("/api/v1.0/Breweries")
def Brewereies():
    return getBreweries_byState()

@app.route("/api/v1.0/State")
def State():
    return getBreweries_byState()

@app.route("/api/v1.0/State/<stateCode>")
def breweriesByState(stateCode):
    return getBreweries_byState(stateCode)

@app.route("/api/v1.0/Cities/<cityName>")
def Cities(cityName):
    return getBreweries_byCity(CityName)

    
@app.route("/api/v1.0/Beers/<beerType>")
def beersByType(beerType):
    return getBreweries_byType(beerType) 


# 4. Define main behavior
if __name__ == "__main__":
       app.run()
    #    app.run(debug=True)
```
